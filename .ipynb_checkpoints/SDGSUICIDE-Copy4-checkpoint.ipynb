```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
from urllib.request import urlopen
import xml.etree.ElementTree as ET
```

```python
resp = requests.get('http://apps.who.int/gho/athena/api/GHO/NCDMORT3070,SDGSUICIDE.xml?filter=COUNTRY:-;REGION:*;SEX:BTSX')
resp.status_code
```

```python
sdg_suicide = resp.content
print(sdg_suicide)
```

```python
with open('data_sdg_suicide', 'wb') as fw:
    fw.write(sdg_suicide)
```

```python
tree = ET.parse('data_sdg_suicide')
tree
```

```python
root = tree.getroot()
```

```python
root[0]
```

```python
for label in root:
    for label2 in label:
        for label3 in label2:
            print(label3.tag, label3.attrib)
```

```python
elements = [label3.tag for label3 in root.iter()]
elements
```

```python
for title in root.iter('Display'):
    print(title.text)
```
